```python
'''
Practical 2: Write a NumPy program to implement following operation 
            
            -> to convert a list of numeric values into a one-dimensional NumPy array 
            -> to create a 3x3 matrix with values ranging from 2 to 10 ï‚· 
            -> to append values at the end of an array 
            -> to create another shape from an array without changing its data(3*2 to 2*3)  
'''

# 2.1 Converting a list of numeric values into a one-dimensional NumPy array 

import numpy as np

# Creating and Printing a list
list_1 = [12,10,8,6,4,2,0]
print("List:  " , list_1, "\n")

# Printing the type for cross checking
print("Type: " ,type(list_1) , "\n")

# Converting and Printing the list into ndarray
arr = np.array(list_1)
print("One-Dimensional Array:  " , arr, "\n")

# Printing the type for cross checking
print("Type:" ,type(arr))
```

```python
# 2.2 Create a 3x3 matrix with values ranging from 2 to 10 

# Creating a matrix having values 2-10
mat1 = np.arange(2,11)

# Converting into 3x3 matrix
mat2 = mat1.reshape(3,3)
print(mat2)
```

```python
# 2.3 Append values at the end of an array

#Creating a list
a1 = [0,0o1,0o2,0o3,0o4]  # Octal numbers are stored therefore preceeded by '0o' .

# Converting List into ndarray
nd_a1 = np.array(a1)
print("Array: " , nd_a1)

#Creating another list
a2 = [0,11,22,33,44]

#Converting above list into ndarray
nd_a2 = np.array(a2)

#Appending Secong array into the first array
new = np.append(nd_a1,nd_a2)
print("\nNew Appended Array: " , new)
```

```python
# 2.4 Creating another shape from an array without changing its data(3*2 to 2*3)

# Creating an ndarray of 3x2 matrix
ar3 = np.array([[1,2],[3,4],[5,6]])
print(" 3x2 matrix: \n" ,ar3)

# Converting 3x2 matrix into 2x3 matrix
ar4 = ar3.reshape(2,3)
print("\n After Conversion\n\n 2x3 matrix:\n", ar4)
```
