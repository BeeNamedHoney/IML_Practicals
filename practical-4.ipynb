```python
'''
Practical 4: Write a NumPy program to implement following operation 
          to add, subtract, multiply, divide arguments element-wise 
          to round elements of the array to the nearest integer 
          to calculate mean across dimension, in a 2D numpy array 
          to calculate the difference between neighboring elements, 
           element wise of a given array 
'''

import numpy as np

# Adding,Subtracting, Multiplying and Dividing Arrays element-wise

#  Defining arrays:
Arr1 = np.array([11,13,15])
Arr2 = np.array([2,4,6])

print(Arr1 , " + " , Arr2 ," = \t" ,np.add(Arr1,Arr2))
print(Arr1 , " - ", Arr2, "=\t " , np.subtract(Arr1,Arr2))
print(Arr1 , " x ", Arr2, "=\t " ,np.multiply(Arr1,Arr2))
print(Arr1 , " % ", Arr2, "=\t " ,np.divide(Arr1,Arr2))
```

```python
# Rounding elements to nearest integers

Arr3 = np.array([1.6,5.5,10.4])
print("Rounding off elements of Array", Arr3 , ": " ,np.round(Arr3))
```

```python
# Calculating mean columnwise and rowwise

Arr4 = np.array([[100,200,300],[400,500,600]])
print(Arr4)
print("\nColumnwise Mean: ", np.mean(Arr4,axis=0))
print("Rowwise Mean: " , np.mean(Arr4,axis=1))
```

```python
# Calculating difference between neighbouring elements

# Array: [[100,200,300],[400,500,600]]
print(Arr4 , "\n\nDifference columnwise:-\n",np.diff(Arr4,axis=0))
print("\nDifference rowwise:-\n",np.diff(Arr4,axis=1))
```
